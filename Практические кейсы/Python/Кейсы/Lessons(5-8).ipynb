```python
''' Вам дан набор значений: нижняя граница диапазона, верхняя граница диапазона и контрольное число. 
Ваша задача — написать цикл, который будет увеличивать значение верхней границы диапазона на 10 до тех пор, пока контрольное число не войдет в диапазон. 
Ответом будет число итераций, за которые мы смогли увеличить верхнюю границу диапазона до контрольного числа.

Нижняя граница диапазона: ln(100) (натуральный логарифм, логарифм по основанию e ≈ 2.7)
Верхняя граница диапазона: квадратный корень из 144
Контрольное значение: логарифм по основанию 2 от 8, возведённый в 4-ю степень'''

import math

lower_border = math.log(100)
upper_border = math.sqrt(144)
control_metric = pow(math.log(8,2), 4)
i = 0

while control_metric > upper_border:
    upper_border += 10
    i += 1

print(lower_border)
print(upper_border)
print(control_metric)
print(i)
```

Мини-кейс №6

```python
import math

table = {'id' : [1,2,3,4,5,6],
         'Стаж в компании (V)' : [24,44,6,67,10,12], 
         'Доходность за 6 месяцев (Y)' : [6500,870,12500,4400,0,3056],
         'Тип компании' : ['ip','ooo','ooo','ip','ip','ooo'],
         'Оборот средств за последние 6 месяцев (R)' : [4500,34500,220000,220,0,15600]}

rate = list()
k = 0 # коэффициент для ооо и ip
for i in range(len(table)):
    if table['Оборот средств за последние 6 месяцев (R)'][i] == 0:
        continue # спящих бизнесменов не берем в учет
    if table['Стаж в компании (V)'][i] < 13 and table['Тип компании'][i] == 'ip':
        k = 0.8 # понижающий кэф для ИП
    elif table['Стаж в компании (V)'][i] < 13 and table['Тип компании'][i] == 'ooo':
        k = 0.9 # понижающий кэф для ООО
    else:
        k = 1
    rate.append(round((math.log(table['Оборот средств за последние 6 месяцев (R)'][i],2)/((table['Стаж в компании (V)'][i]/table['Доходность за 6 месяцев (Y)'][i])+0.1))*k,2))

table['Рейтинг привлекательности'] = rate
print(table)
```
