```python
compa = {'id': [1,2], 'type_': ['ИП','ООО'], 'region': ['Москва', 'Ярославль'], 'revenue': [400000, 300000], 'age_month': [15, 11]}

id_num = int(input()) - 1

if ((compa['type_'][id_num] == 'ИП' and compa['revenue'][id_num] > 300000) or (compa['type_'][id_num] == 'ООО' and compa['revenue'][id_num] > 600000)) and (compa['region'][id_num] == 'Москва' and compa['age_month'][id_num] >= 12):
    print('Первый приоритет')
elif ((compa['type_'][id_num] == 'ИП' and compa['revenue'][id_num] > 100000) or (compa['type_'][id_num] == 'ООО' and compa['revenue'][id_num] > 300000)) and (compa['region'][id_num] == 'Москва' and compa['age_month'][id_num] >= 6):
    print('Второй приоритет')
else:
    print('Третий приоритет')
```
